```python
from sentence_transformers import SentenceTransformer
import pandas as pd
import numpy as np
import pickle

# Load your dataset
df = pd.read_csv("cleaned_disease_dataset.csv")

# Ensure the sentence column is in string format
sentences = df['Generated_Sentence_From_symptoms'].astype(str).tolist()

# Load the transformer model (you can switch to another if needed)
model = SentenceTransformer('all-MiniLM-L6-v2')

# Generate embeddings for all sentences
embeddings = model.encode(sentences, show_progress_bar=True)

# Save the embeddings, disease names, and source URLs
data = {
    'embeddings': embeddings,
    'disease_names': df['Disease_Name'].tolist(),
    'source_urls': df['Source_URL'].tolist(),
    'sentences': sentences
}

# Save everything to a file for later use during inference
with open('transformer_model.pkl', 'wb') as f:
    pickle.dump(data, f)

print("✅ Embeddings generated and saved to disease_embeddings.pkl")
```
