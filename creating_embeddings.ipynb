```python
import pandas as pd

# Load your cleaned dataset (this file should already be created)
df = pd.read_csv("cleaned_disease_dataset.csv")
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

# Step 1: Load the pre-trained transformer model
transformer_model = SentenceTransformer('all-MiniLM-L6-v2')  # You can choose a different model if needed

# Step 2: Generate embeddings for all sentences in the 'Generated_Sentence_From_symptoms' column
embeddings = transformer_model.encode(df['Generated_Sentence_From_symptoms'].tolist())

# Step 3: Save the embeddings to a file for later use
np.save('embeddings.npy', embeddings)

# Now you have the embeddings saved and can use them for comparison later
```
